```python
import pandas as pd 
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
plt.style.use('ggplot')
```

```python
ts = pd.Series(np.random.randn(1000), index = pd.date_range('1/2/2000',periods=1000))
```

```python
ts = ts.cumsum()
```

```python
ts.plot(title = 'a picture')
plt.xlabel('xdata')
plt.ylabel('ydata')
```

```python
df = pd.DataFrame(np.random.randn(1000,4), index = pd.date_range('1/2/2000',periods=1000),columns=list('ABCD'))
```

```python
df.cumsum()
```

```python
df['A'] = np.arange(1000)
df.plot(x='A',y='C')
```

```python
df.C[:5].plot.bar(rot=20,color='g')
plt.axhline(0,color='r')
```

```python
df.C.plot.hist(bins=50)
```

```python
df.plot.scatter(x='D',y='B')
```

```python
np.abs(df.D[:6]).plot.pie(figsize=(10,10)) 
```
