```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
from sklearn import metrics
%matplotlib inline
```

```python
data1 = pd.DataFrame({'X':np.random.randint(1,50,100), 
                      'Y':np.random.randint(1,50,100)})
data = pd.concat([data1 + 50, data1])
```

```python
plt.style.use('ggplot')
plt.scatter(data.X, data.Y)
```

```python
y_pred = KMeans(n_clusters=2).fit_predict(data)
```

```python
plt.scatter(data.X, data.Y, c=y_pred)
```

```python
# 评价结果，值越大越好
metrics.calinski_harabaz_score(data, y_pred)
```
