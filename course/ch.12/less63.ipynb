```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN
```

```python
data1 = pd.read_csv('../fangchan/lianjia1.csv', encoding='gbk')
data2 = pd.read_csv('../fangchan/lianjia2.csv', encoding='gbk')
data3 = pd.read_csv('../fangchan/lianjia3.csv', encoding='utf-8')
data4 = pd.read_csv('../fangchan/lianjia4.csv', encoding='utf-8')
data5 = pd.read_csv('../fangchan/lianjia5.csv', encoding='utf-8')
data6 = pd.read_csv('../fangchan/lianjia6.csv', encoding='utf-8')
data7 = pd.read_csv('../fangchan/lianjia7.csv', encoding='utf-8')
data = pd.concat([data1, data2, data3, data4, data5, data6, data7])
data = data.dropna()
```

```python
data['cjdanjia'] = data.cjdanjia.str.replace('元/平', '').astype(np.float32).map(lambda x: round(x/10000, 2))
```

```python
data1= data[data.cjxiaoqu.str.contains('龙锦苑东一区 3室2厅 124平')]
```

```python
plt.scatter(range(1, len(data1) + 1), data1.cjdanjia)
```

```python
y_pred = KMeans(n_clusters=2).fit_predict(data1[['cjdanjia']])
```

```python
plt.scatter(range(1, len(data1) + 1), data1.cjdanjia, c=y_pred)
```

```python
data2 = data[data.cjxiaoqu.str.contains('龙锦苑东五区 3室2厅 124平')]
y_pred2 = KMeans(n_clusters=3).fit_predict(data2[['cjdanjia']])
plt.scatter(range(1, len(data2) + 1), data2.cjdanjia, c=y_pred2)
```

DBSCAN 根据密度分类

```python
y_pred = DBSCAN(eps=0.4).fit_predict(data2.cjdanjia.values.reshape(-1, 1))
```

```python
plt.scatter(range(1, len(data2) + 1), data2.cjdanjia, c=y_pred)
```
